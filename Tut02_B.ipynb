```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data = pd.read_csv("https://tinyurl.com/ChrisCoDV/world_population.csv", index_col =0)

```

```
data = data.transpose()
selected = ['Country Code','Indicator Name','Indicator Code']
data = data.drop(selected, axis = 0)
print(data)
print(data.describe())
```

From above, it is shown that the investigation was carried out over 57 years with 264 countries population density from 1960 to 2016.

```
data = data.reindex(data.mean().sort_values(ascending = False).index, axis= 1)
plt.figure(figsize=(8,8))
x_pos = np.arange(len(data.columns))
plt.bar(x_pos, data.mean(),align='center')
plt.xticks(x_pos,data.columns)
plt.xlabel('Country Name', fontsize=18)
plt.ylabel('Population (per square kilometer)', fontsize = 18)
plt.title (' Mean Population density', fontsize=20)
plt.show()

```

```
categories = ['Extreme', 'High','Moderate', 'Low']
categories_selected = [[] for i in range(len(categories))]
for name in data.columns:
    average_density = data[name].mean()
    if average_density  > 6000:
        category = 0
    elif average_density > 2000:
        category = 1
    elif average_density  > 1000:
        category = 2
    else:
        category = 3
    categories_selected[category].append(name)
    print('Country ' + name + ' is ' + categories[category] + ' population')

for i in range(len(categories)):
    print(categories[i] + ': ' + str(categories_selected[i]))

for i, selected in enumerate(categories_selected):
    plt.figure(figsize=(8, 8))
    x_pos = np.arange(len(data[selected].columns))
    plt.bar(x_pos, data[selected].mean(), align='center')
    plt.xticks(x_pos, data[selected].columns)
    plt.xlabel('Country', fontsize=18)
    plt.ylabel('Average Population (per square kilometer)', fontsize=18)
    plt.title(categories[i] + ' Average Population Density', fontsize=20)
    plt.show()
```

```
selected = ['Libya', 'Bermuda', 'Macao SAR, China','Singapore','Ireland']
data_selected = data[selected]

plt.figure(figsize=(8,8))
x_pos = np.arange(len(data_selected.columns))
plt.bar(x_pos,data_selected.mean(), align= 'center')
plt.xticks(x_pos,data_selected.columns)
plt.xlabel('Country', fontsize = 18)
plt.ylabel('Average population (per square kilometer)', fontsize = 18)
plt.ylim(ymax=20000)
plt.title('Selected Countries Average Population', fontsize = 20)
plt.show()
```

```
year_selected = data_selected.iloc[1:11, ]
print(year_selected)

plt.figure(figsize=(8,8))
x_pos = np.arange(len(year_selected.columns))
plt.bar(x_pos,year_selected.mean(), align= 'center')
plt.xticks(x_pos,year_selected.columns)
plt.xlabel('Country', fontsize = 18)
plt.ylabel('Average population(per square kilometer)', fontsize = 18)
plt.ylim(ymax=20000)
plt.title('Selected Countries Average Population from 1961 to 1970', fontsize = 20)
plt.show()
```
